Найти все простые числа меньше или равные заданному числу N.


```python
import numpy as np

def find_primes(n):
    if n < 2:
        return []
    is_prime = [True] * (n + 1) #создаём список из n+1 элементов True - простых чисел
    is_prime[0] = is_prime[1] = False
    p = 2
    while p * p <= n: #проверка делителей
        if is_prime[p]:
            for i in range(p * p, n + 1, p): #генерация кратных чисел p
                is_prime[i] = False
        p += 1
    return [i for i in range(2, n + 1) if is_prime[i]]

try:
    N = int(input("Введите число N: "))
    if N < 0:
        print("Введите неотрицательное число.")
    else:
        primes = find_primes(N)
        print(f"Простые числа ≤ {N}: {primes}")
        print(f"Количество простых чисел: {len(primes)}")
except ValueError:
    print("Ошибка: введите целое число.")
```

Решето Эратосфена:
  
  Преимущество: быстрота - эффективнно для больших N.

1. создаем массив чисел is_prime от 0 до N, изначально помеченных как простые (True);
2. 0 и 1 не являются простыми,помечены как False;
3. начиная с 2, если число простое, помечаем все его кратные как непростые, начиная с $p^2$ как составные;
4. цикл **while** работает до тех пор, пока текущее число не станет больше $\sqrt{N}$ - sqrt(N) (потому что все составные числа до N будут помечены кратными простым числам, не превосходящих $\sqrt{N}$).  
6. cобираем все числа, где is_prime[i] = True


Временная сложность O(n log log n)  
Если N = 30
O(N logN logN) = 30 * log(log(30)) = 36.6

Задача 2.  
Определить, будет ли побитовое ИЛИ чисел в заданном массиве чётным или нечётным числом.

```python
def is_bitwise_or_even(arr):
    for num in arr:
        if num % 2 == 1:  #проверяем  младший бит на нечетность # 50 % 2 = 0 - False
            return "Нечётное"
    return "Чётное"


def main():
  array = input("Введите числа массива через пробел: ")
  result = is_bitwise_or_even(list(map(int, array.strip().split())))
  print(f"Побитовое ИЛИ массива {list(map(int, array.strip().split()))}: {result}")

if __name__ == "__main__":
    main()

```



1.   Побитовое ИЛИ является нечётным, если хотя бы одно число в массиве имеет  младший бит.
2. Если все числа чётные (т. е. младший бит = 0), то результат будет чётным.  




Временная сложность O(n):


1.   В лучшем случае - первое число нечетное, функция завершится после первой же проверки O(1)
2. Если же все числа четные, функция пройдется по всем элементам массива.

Пример ввода:  50 10 2   
num=50 - четное,   50 % 2 = 0 - False  
num=10 - четное,  10 % 2 = 0 - False  
num=2 - четное, 2 % 2 = 0 - False  
выходим из цикла и возвращаем "чётное"  
кол.-во проверок: 3
